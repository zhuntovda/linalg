__1.__ Решить систему уравнений методом Гаусса:

$$\begin{cases}
x_{1}+x_{2}-x_{3}-2x_{4}=0, \\
2x_{1}+x_{2}-x_{3}+x_{4}=-2, \\
x_{1}+x_{2}-3x_{3}+x_{4}=4.
\end{cases}$$

$$\tilde{A}=\begin{pmatrix}
\left.\begin{matrix}
1 & 1 & -1 & -2\\ 
2 & 1 & -1 & 1\\ 
1 & 1 & -3 & 1
\end{matrix}\right|
\begin{matrix}
0\\ 
-2\\
4
\end{matrix}
\end{pmatrix} = \begin{pmatrix}
\left.\begin{matrix}
1 & 1 & -1 & -2\\ 
0 & -1 & 1 & 5\\ 
0 & 0 & -2 & 3
\end{matrix}\right|
\begin{matrix}
0\\ 
-2\\
4
\end{matrix}
\end{pmatrix}$$

$$\begin{cases}
x_{1}+x_{2}-x_{3}-2x_{4}=0, \\
-x_{2}+x_{3}+5x_{4}=-2, \\
-2x_{3}+3x_{4}=4.
\end{cases}$$

$$x_{4}=c,$$
$$-2x_{3}+3c=4 \; \Leftrightarrow \; x_{3}=(3c-4)/2,$$
$$x_{1} + (c+2)/3 - c=0 \; \Leftrightarrow \; x_{1}=(2c-2)/3.$$

Таковым будет _общее решение системы_. Подставляя произвольные числа вместо $c$, мы получим _частное решение_. Например, при $c=0$:

$$x_{1}=-2/3,$$
$$x_{2}=2/3,$$
$$x_{3}=0.$$

__4.__ Дана система линейных уравнений, заданная расширенной матрицей

$$\tilde{A}=\begin{pmatrix}
\left.\begin{matrix}
1 & 2 & 3\\ 
4 & 5 & 6\\ 
7 & 8 & 9
\end{matrix}\right|
\begin{matrix}
a\\ 
b\\
c
\end{matrix}
\end{pmatrix}.$$

Найти соотношение между параметрами $a$, $b$ и $c$, при которых система является несовместной.


Так как система является не совместимой если $rank A<>rank \tilde A $, требуется найти соотношение между параметрами $a$, $b$ и $c$ при котором выполняется данное условие.

$\begin{cases}
x_{1}+2x_{2}+3x_{3}=a, \\
4x_{1}+5x_{2}+6x_{3}=b, \\
7x_{1}+8x_{2}+9x_{3}=c;
\end{cases}$

Каждое из данных уровнений, можно выразит через уровнение $(3n-2)x_{1}+(3n-1)x_{2}+3nx_{3}=d$, где $d=a$ при $n=1$ и т.д.
Следовательно, система будет совместной если $a+1=b, b+1=c$ или $a=b=c=0$, в остальных случаях система будет не совместной
